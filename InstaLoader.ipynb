```python
import instaloader
import re
import json

def fetch_post_details(username, password, post_urls, output_file="instagram_posts.json"):
    """
    Fetch details of Instagram posts from the given list of URLs and save to a JSON file.

    Args:
        username (str): Instagram username.
        password (str): Instagram password.
        post_urls (list): List of Instagram post URLs.
        output_file (str): Name of the JSON file to save the data.

    Returns:
        None
    """
    # Initialize Instaloader
    L = instaloader.Instaloader()

    try:
        # Login to Instagram
        print("Logging in to Instagram...")
        L.login(username, password)
        print("Login successful!")

        # Process each post URL
        post_details = []
        for post_url in post_urls:
            try:
                # Extract shortcode from the URL
                shortcode = post_url.strip("/").split("/")[-1]

                # Fetch post using the shortcode
                post = instaloader.Post.from_shortcode(L.context, shortcode)

                # Collect post details
                details = {
                    "post_url": post_url,
                    "post_owner": post.owner_username,
                    "caption": post.caption,
                    "hashtags": post.caption_hashtags,
                    "is_sponsored": "sponsored" in (post.caption or "").lower(),
                    "likes": post.likes,
                    "comments": post.comments,
                }

                post_details.append(details)
                print(f"Fetched details for: {post_url}")

            except Exception as e:
                print(f"Error fetching post {post_url}: {str(e)}")

        # Save the details to a JSON file
        with open(output_file, "w", encoding="utf-8") as json_file:
            json.dump(post_details, json_file, indent=4, ensure_ascii=False)
            print(f"Data saved to {output_file}")

    except instaloader.exceptions.BadCredentialsException:
        print("Invalid credentials. Please check your username and password.")
    except Exception as e:
        print(f"An unexpected error occurred: {str(e)}")

if __name__ == "__main__":
    # Instagram credentials
    username = "****"  # Replace with Instagram username
    password = "****"  # Replace with Instagram password

    # List of Instagram post URLs
    post_urls = [
-
    ]

    # Fetch post details and save to a file
    fetch_post_details(username, password, post_urls)
```

```python
import instaloader
import re

def fetch_post_details(username, password, post_urls):
    """
    Fetch details of Instagram posts from the given list of URLs.

    Args:
        username (str): Instagram username.
        password (str): Instagram password.
        post_urls (list): List of Instagram post URLs.

    Returns:
        list: List of dictionaries containing post details.
    """
    # Initialize Instaloader
    L = instaloader.Instaloader()

    try:
        # Login to Instagram
        print("Logging in to Instagram...")
        L.login(username, password)
        print("Login successful!")

        # Process each post URL
        post_details = []
        for post_url in post_urls:
            try:
                # Extract shortcode from the URL
                shortcode = post_url.strip("/").split("/")[-1]

                # Fetch post using the shortcode
                post = instaloader.Post.from_shortcode(L.context, shortcode)

                # Collect post details
                details = {
                    "post_url": post_url,
                    "video_url": post.video_url if post.is_video else None,
                    "caption": post.caption,
                    "hashtags": post.caption_hashtags,
                    "is_sponsored": "sponsored" in (post.caption or "").lower(),
                    "likes": post.likes,
                    "comments": post.comments,
                }

                post_details.append(details)
                print(f"Fetched details for: {post_url}")

            except Exception as e:
                print(f"Error fetching post {post_url}: {str(e)}")

        return post_details

    except instaloader.exceptions.BadCredentialsException:
        print("Invalid credentials. Please check your username and password.")
        return []
    except Exception as e:
        print(f"An unexpected error occurred: {str(e)}")
        return []

if __name__ == "__main__":
    # Instagram credentials
    username = "****"  # Replace with your Instagram username
    password = "****"  # Replace with your Instagram password

    # List of Instagram post URLs
    post_urls = [
        "https://www.instagram.com/p/DEiK6IyI8xE/",
        "https://www.instagram.com/p/DEVXCo-oFzN",
        # Add more post URLs here
    ]

    # Fetch post details
    posts = fetch_post_details(username, password, post_urls)

    # Print the details
    if posts:
        for post in posts:
            print("Post Details:")
            for key, value in post.items():
                print(f"  {key}: {value}")
            print("-" * 50)
    else:
        print("No data found or an error occurred.")
```
